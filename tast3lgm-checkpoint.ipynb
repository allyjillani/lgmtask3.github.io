```python
from keras.datasets import mnist
```

```python
objects=mnist
```

```python
(train_dt, train_lbl),(test_dt, test_lbl)=objects.load_data()
```

```python
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(train_dt[i], cmap="gray_r")
    plt.title("digits :{}".format(train_lbl[i]))
    plt.subplots_adjust(hspace=0.3)
    plt.axis("off")
```

```python
train_dt=train_dt/255.0
```

```python
test_dt=test_dt/255.0
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten

model=Sequential()
input_layer = Flatten(input_shape=(28,28))
model.add(input_layer)
hidden1_layer=Dense(512, activation="relu")
model.add(hidden1_layer)
hidden2_layer=Dense(512, activation="relu")
model.add(hidden2_layer)
out_layer=Dense(512, activation="softmax")
model.add(out_layer)
```

```python
model.compile(optimizer="adam",
             loss="sparse_categorical_crossentropy",
             metrics=['accuracy'])
```

```python
model.fit(train_dt, train_lbl, epochs=25)
```

```python
loss_auc=model.evaluate(test_dt, test_lbl, verbose=2)
```
